```python
price=df["Price"]
```

```python
import matplotlib.pyplot as plt
```

```python
print(min(price))
print(max(price))
print(max(price)-min(price) /10)
```

```python
fig=plt.figure()

plt.hist(x=price ,edgecolor="yellow" ,bins=20000)
plt.xlabel("Price range")
plt.ylabel("Number of houses")
plt.title("Histogram of price")
plt.show()
```

```python
rooms_count=df["Regionname"].value_counts()
```

```python
print(rooms_count)
```

```python
rooms_count=dict(rooms_count)
```

```python
print(rooms_count)
```

```python
a=rooms_count.keys()
b=rooms_count.values()
```

```python
#pie
fig=plt.figure(figsize=(50,50)) 
plt.pie(x=b, labels=a)

plt.legend()

plt.show()
```

```python
type_count=df["Type"].value_counts()
type_count=dict(type_count)
print(type_count)
```

```python
k=list(type_count.keys())
v=list(type_count.values())
```

```python
fig=plt.figure(figsize=(10,10))
plt.bar(x=k,height=v,edgecolor="red",color="Yellow") 
plt.show()
```

```python
price=df["Price"]
price=list(price)
distance=df["Distance"]
distance=list(distance)
```

```python
fig=plt.figure(figsize=(5,5))
#lineplot
plt.plot(distance,price, linestyle="",marker="o",
        markerfacecolor="red",
        markeredgecolor="yellow")

plt.xlabel("Distance")
plt.ylabel("Price")
plt.title("Distance vs Price")
plt.show()
```
